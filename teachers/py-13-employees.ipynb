### Read all other DB tables

```python
tables = ['Categories', 'CustomerCustomerDemo', 'CustomerDemographics', 'Customers',
          'EmployeeTerritories', 'OrderDetails', 'Orders', 'Products', 'Region',
          'Shippers', 'Suppliers', 'Territories', 'invoices',
          'purchase_order_status', 'purchase_orders', 'purchase_order_details']
```

## Compute the number of employees hired in each year

```python
employees['HireDate'] = pd.to_datetime(employees['HireDate'])
```

```python
employees['year'] = employees['HireDate'].dt.year
```

```python
employees.head()
```

```python
employees.groupby('year').size()
```

## Find the manager (that is, employee that do not report to any other person), who have been hired the most time.

First, select the managers

```python
managers = employees[employees['ReportsTo'].isnull()]
managers.head()
```

Then we can use `idxmin` to extract the manager with the earliest `HireDate`

```python
managers['HireDate'].idxmin()
```

```python
managers.loc[managers['HireDate'].idxmin()]
```

## Find the orders with the longest time between the submission and the approval.

The table `purchase_orders` has columns `submitted_date` and `approved_date`. If any date is missing, we assume the order has not been approved.

```python
df['purchase_orders']
```

```python
approved = df['purchase_orders'][df['purchase_orders']['approved_date'].notnull()].copy()
approved.head()
```

```python
approved['duration'] = approved['approved_date'] - approved['submitted_date']
approved.head()
```

Then we can use `idxmax` to extract the manager with the larges duration

```python
approved.loc[approved['duration'].idxmax()]
```

```python
approved[approved['duration'] >= approved['duration'].max()]
```

## Find the order that has the most items

```python
df['OrderDetails'].columns
```

```python
items = df['OrderDetails'].groupby('OrderId', as_index = False)['Quantity'].sum()
items
```

```python
max_items = items[items['Quantity'] >= items['Quantity'].max()]
max_items
```

```python
pd.merge(max_items, df['Orders'], left_on = 'OrderId', right_on = 'Id')
```

## Compute how the current number of units in stock is distributed among all categories

```python
q6 = df['Products'].groupby('CategoryId', as_index = False)['UnitsInStock'].sum()
q6
```

## Find the categories with the largest number of items in stock.

```python
q6[q6['UnitsInStock'] >= q6['UnitsInStock'].max()]
```
