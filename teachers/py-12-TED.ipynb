## Compute the number of talks given in each year (film_date)

```python
ted = pd.read_csv('https://github.com/gdv/foundationsCS/raw/main/students/ex-data/TED/ted_main.csv') 
```

```python
ted.head()
```

```python
def get_year(timestamp):
    return pd.to_datetime(timestamp, unit='s').year
get_year(1140825600)
```

```python
ted["year"] = ted["film_date"].apply(get_year)
# Alternative with lambda function
ted["year"] = ted["film_date"].apply(lambda x: pd.to_datetime(x, unit='s').year)
```

```python
ted.head()
```

```python
ted.groupby('year').size()
```

## How many talks have been given by a person with only one occupation?

Hint: assume that a missing occupation means that the number of occupations is zero.

```python
import re
import numpy as np
def num_jobs(s):
    if s is np.NaN or s == '':
        return 0
    else:
        return len(re.findall("[,;]", str(s))) + 1
```

```python
ted['num_occupation'] = ted['speaker_occupation'].apply(num_jobs)
ted.head()
```

```python
len(ted[ted['num_occupation'] == 1])
```
