```python
import pandas as pd
import datetime
import sqlite3
```

First we need to connect to a database. Only after the connection is established, we can query the DB.

We cannot use to URL, but we need to download the [DB](https://github.com/gdv/foundationsCS/raw/master/students/ex-data/Northwind/Northwind_large.sqlite) and save it on the disk.

```python
cnx = sqlite3.connect('../students/ex-data/Northwind/Northwind_large.sqlite')
c = cnx.cursor()
```

```python
c.execute('SELECT * FROM employees')
results = c.fetchall()
results
```

```python
employees = pd.DataFrame(results)
employees.head()
```

```python
employees = pd.read_sql_query('SELECT * FROM employees', cnx)
employees.head()
```
