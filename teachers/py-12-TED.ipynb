## Compute the number of talks given in each year (film_date)

```python
ted = pd.read_csv('https://github.com/gdv/foundationsCS/raw/main/students/ex-data/TED/ted_main.csv') 
```

```python
ted.head()
```

We will build a function `get_year` that extracts the year of any timestamp. We check that it is applied only to 21st century dates.

```python
def get_year(timestamp):
    return pd.to_datetime(timestamp, unit='s').year
get_year(1140739200)
```

```python
ted['year'] = ted['film_date'].apply(get_year)
#assert ted['year'].min() >= 2000, "date too early"
#assert ted['year'].max() <= 2018, "date too late"
ted.head()
```

```python
ted[ted['year'] == 1972]
```

The previous tests show that it is too restrictive to assume that all `film_date`s are from the 21st century. We can test if the same test holds fro `published_date` 

```python
ted['published_date'].apply(get_year).min()
```

```python
ted.groupby('year').size()
```

## Find all talks with tag *MacArthur grant* (they can have also other tags)

Hint: the `tags` column contains a JSON list. Remember that a JSON list is also in Python format.
Therefore we can use the `ast` library, but several other ways are possible.

```python
import ast
def tags_list(tag):
    return "MacArthur grant" in ast.literal_eval(tag)

ted[ted['tags'].apply(tags_list)]
```

## How many talks have been given by a person with only one occupation?

```python
ted[ted['speaker_occupation'].isnull()]
```

Hint: assume that a missing occupation means that the number of occupations is zero.

```python
import re
import numpy as np

def num_jobs(s):
    if s is np.NaN or s == '':
        return 0
    else:
        return len(re.split("[,;]", str(s)))
```

```python
ted['num_occupation'] = ted['speaker_occupation'].apply(num_jobs)
```

```python
len(ted[ted['num_occupation'] == 1])
```
